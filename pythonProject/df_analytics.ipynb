```python
def extract_children_dep(pattern: str, 
                         sentence: str, 
                         lowercase=True) -> str:
    
    sentence = sentence.lower() if lowercase else sentence
    sentence_proccessed = nlp(sentence)
    list_sent_proc = list(sentence_proccessed)
    is_nec_word = lambda tok: re.fullmatch(pattern, str(tok), flags=re.IGNORECASE)
    all_tokens = '; '.join(re.findall(pattern, sentence, flags=re.IGNORECASE))

    nec_tokens = [(inx, word) for inx, word in enumerate(sentence_proccessed) 
                  if (nec_word := is_nec_word(word)) 
                  and word.pos_ == 'NOUN'] 
    children_groups = list(list(token[-1].children) + [';'] for token in nec_tokens)
    nec_children_flatten = list(filter(lambda x: isinstance(x, str) or x.pos_ != 'PUNCT', chain(*children_groups)))
    children_flatten_filtered = list(filter(lambda word: not isinstance(word, str), nec_children_flatten))
    nec_tokens += list((inx, word) for inx, word in enumerate(sentence_proccessed) 
                       if word in children_flatten_filtered)
    try:
        for inx in range(len(nec_children_flatten)):
            child = nec_children_flatten[inx]
            if isinstance(child, str) and child == ';':
                delimeter_inx = list_sent_proc.index(nec_children_flatten[inx - 1]) + 1
                nec_tokens.append((delimeter_inx, child))
    except TypeError:
        pass

    restored_word_order = [str(pair[-1]) for pair in sorted(nec_tokens, key=lambda x: x[0])]
    restored_word_order = ["".join(restored_word_order[inx - 2: inx]) 
                           if restored_word_order[inx - 1] == ';'
                           else restored_word_order[inx - 1]
                           for inx in range(1, len(restored_word_order))] # блять индексы
    phrase_concatenated = ' '.join(restored_word_order)
    return phrase_concatenated if phrase_concatenated else all_tokens if all_tokens else np.nan
```

```python
sim_dict = {}

def calculate_similarity(nec_lemma: str, 
                         sentence: str) -> dict: 
    nec_token, doc = nlp(nec_lemma)[0], nlp(sentence)
    noun_span_list = [nlp(token.lemma_)[0] for token in doc if
                      token.pos_ == 'NOUN' and str(token.lemma_) != 'правда']
    for token in noun_span_list:
        string_tok = str(token)
        if string_tok not in sim_dict:
            sim_dict.update({string_tok: nec_token.similarity(token)})
```
